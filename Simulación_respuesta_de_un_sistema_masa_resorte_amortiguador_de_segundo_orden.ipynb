```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Función del sistema masa-resorte-amortiguador
def mass_spring_damper(t, y, m, b, k):
    x, v = y
    dxdt = v
    dvdt = (-b * v - k * x) / m
    return [dxdt, dvdt]

# Parámetros del sistema
params = {
    "masa": 500,      # kg
    "amortiguador": 100,  # Ns/m
    "resorte": 200   # N/m
}

# Condiciones iniciales
x0 = 2  # Posición inicial
v0 = 0  # Velocidad inicial

# Rango de tiempo
t_span = [0, 10]
t_eval = np.linspace(0, 10, 100)

# Solución numérica
sol = solve_ivp(mass_spring_damper, t_span, [x0, v0], args=(params["masa"], params["amortiguador"], params["resorte"]), t_eval=t_eval)

# Mostrar datos de entrada
print("Datos de Entrada:")
print(f"Masa: {params['masa']} kg")
print(f"Amortiguador: {params['amortiguador']} Ns/m")
print(f"Resorte: {params['resorte']} N/m")
print(f"Condiciones iniciales: x0 = {x0}, v0 = {v0}\n")

# Mostrar datos de salida: (primeros 10 valores)
print("Datos de Salida (Primeros 10 puntos)")
for i in range(10):
    print(f"Tiempo: {sol.t[i]:.2f} s, Posición: {sol.y[0, i]:.4f} m, Velocidad: {sol.y[1, i]:.4f} m/s")

# Color personalizable
color_grafica = "red"

# Graficar la solución
plt.plot(sol.t, sol.y[0], color=color_grafica, label="Posición")
plt.title("Sistema Masa-Resorte-Amortiguador")
plt.xlabel("Tiempo (s)")
plt.ylabel("Posición (m)")
plt.legend()
plt.grid()
plt.show()
```
